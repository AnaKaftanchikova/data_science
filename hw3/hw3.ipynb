```python
import data_loader
import data_analyze
import data_visual
import database
import log
```

```python
"""
dataset from https://www.kaggle.com/datasets/vivekattri/california-wildfire-damage-2014-feb2025
"""
file_path = "California Wildfire Damage.csv"
```

```python
# открытие файла CSV
contents = data_loader.open_file(file_path)
```

```python
db = database.SQLiteDB('fires.db')

# определение структуры таблицы
columns = {
    'id': 'INTEGER PRIMARY KEY AUTOINCREMENT',
    'id_incident': 'TEXT NOT NULL',
    'date_fire': 'DATE NOT NULL',
    'location': 'TEXT NOT NULL',
    'area_burned': 'INT NOT NULL',
    'homes_destroyed': 'INT NOT NULL',
    'businesses_destroyed': 'INT NOT NULL',
    'vehicles_damaged': 'INT NOT NULL',
    'injuries': 'INT NOT NULL',
    'fatalities': 'INT NOT NULL',
    'estimated_financial_loss': 'REAL NOT NULL',
    'cause': 'TEXT NOT NULL'
}

# создание таблицы
db.create_table('fires_incidents', columns)
```

```python
# вставка данных
columns_insert = '''
    id_incident,
    date_fire,
    location,
    area_burned,
    homes_destroyed,
    businesses_destroyed,
    vehicles_damaged,
    injuries,
    fatalities,
    estimated_financial_loss,
    cause
'''
columns_values_insert = '?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?'

db.insert_data_many('fires_incidents', columns_insert, columns_values_insert, contents)
```

```python
# получение и вывод всех данных
fires = db.fetch_all('fires_incidents')
print("Данные в таблице fires_incidents:")
for res in fires:
    print(res)
```

```python
# 1-ая строка - заголовки, поэтому удалим ее
db.delete_data("DELETE FROM fires_incidents WHERE Cause = 'Cause'")
fires = db.fetch_all('fires_incidents')
print("Данные в таблице fires_incidents:")
for res in fires:
    print(res)
```

```python
# запросы с WHERE
fires = db.fetch_data_with_query("SELECT * FROM fires_incidents WHERE Cause = 'Lightning'")
print("Данные в таблице fires_incidents:")
for res in fires:
    print(res)
```

```python
# запросы с агрегатными функциями
fires = db.fetch_data_with_query("SELECT Cause, COUNT(*) FROM fires_incidents GROUP BY Cause ORDER BY Cause")
print("Данные в таблице fires_incidents:")
for res in fires:
    print(res)
```

```python
data = db.prepare_data("SELECT cause, COUNT(*) as cnt_incidents FROM fires_incidents GROUP BY Cause ORDER BY Cause")
print(data)

try:
    figure = data_visual.add_line_plot(data, 'cause', 'cnt_incidents', 'cnt_cause')
    log.add_log_info('Построение столбчатой диаграммы отработало')
except Exception as e:
    log.add_log_error(f"Произошла ошибка: {e}")
else:
    log.add_log_info('Сохранение столбчатой диаграммы')
finally:
    log.add_log_debug('finally_after_add_bar')
```

```python
data = db.prepare_data('''
    WITH CTE AS (
        SELECT date(date_fire, 'start of year') as date_fire, COUNT(*) as cnt_rows 
        FROM fires_incidents GROUP BY date_fire ORDER BY date_fire
        )
    SELECT substr(date_fire,1,4) as date_fire, SUM(cnt_rows) as cnt_incidents 
    FROM CTE 
    GROUP BY date_fire 
    ORDER BY date_fire
    ''')
print(data)

try:
    figure = data_visual.add_plot(data, 'date_fire', 'cnt_incidents', 'year_incidents')
    log.add_log_info('Построение линейного графика отработало')
except Exception as e:
    log.add_log_error(f"Произошла ошибка: {e}")
else:
    log.add_log_info('Сохранение линейного графика')
finally:
    log.add_log_debug('finally_after_add_plot')
```

```python
data = db.prepare_data('''
    WITH CTE AS (
        SELECT date(date_fire, 'start of year') as date_fire, cause, sum(estimated_financial_loss) as cnt_rows 
        FROM fires_incidents 
        GROUP BY date_fire, cause
        )
    SELECT substr(date_fire,1,4) as date_fire, cause, SUM(cnt_rows) as sum_loss 
    FROM CTE 
    GROUP BY date_fire, cause 
    ORDER BY date_fire
    ''')
print(data)

try:
    figure = data_visual.add_lineplot(data, 'date_fire', 'sum_loss', 'cause_loss', 'cause')
    log.add_log_info('Построение предполагаемого финансового ущерба отработало')
except Exception as e:
    log.add_log_error(f"Произошла ошибка: {e}")
else:
    log.add_log_info('Сохранение предполагаемого финансового ущерба')
finally:
    log.add_log_debug('finally_after_add_lineplot')
```
